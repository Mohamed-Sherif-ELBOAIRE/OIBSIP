- On the evening of 24 March 2020, the Government of India ordered a nationwide lockdown 
- So, we are going to Analyze the differences upon unemployment rate before and after April 2020
- The first unlock came with a fresh set of guidelines for the month of June. 

```python
before_lockdown = df[(df['Month'] >= 1) & (df['Month'] <4)]
lockdown = df[(df['Month'] >= 4) & (df['Month'] <=7)]
after_lockdown = df[(df['Month'] >7) & (df['Month'] <=12)]
```

```python
lockdown = lockdown.groupby('States')['Estimated Unemployment Rate'].mean().reset_index()

bf_lock = before_lockdown.groupby('States')['Estimated Unemployment Rate'].mean().reset_index()
bf_lock['unemployment rate during lockdown'] = lockdown['Estimated Unemployment Rate']

af_lock = after_lockdown.groupby('States')['Estimated Unemployment Rate'].mean().reset_index()
bf_lock['Estimated Unemployment Rate After Lockdown'] = af_lock['Estimated Unemployment Rate']


bf_lock.columns = ['States','Estimated Unemployment Rate Before Lockdown','Estimated Unemployment Rate During Lockdown','Estimated Unemployment Rate After Lockdown']
round(bf_lock,2)
```

```python
# Performing a Data Transformation , Reshape the Data to Visualize with Animation Frame
melted_bf_lock = bf_lock.melt(id_vars=['States'], # States that will remain as identifiers and won't be melted
                              value_vars=['Estimated Unemployment Rate Before Lockdown',
                                          'Estimated Unemployment Rate During Lockdown', 
                                          'Estimated Unemployment Rate After Lockdown'],
                              # These columns will melt into a single column make the df with 3 columns one for States,
                              # and duration time of Unemployment Rate and the last one for the percentage rate
                              var_name='Period', # Assign new Column Name for Durations
                              value_name='Unemployment Rate') # Assign new Column Name for Values

# Create the animated bar plot
fig = px.bar(
    melted_bf_lock,
    x='States',
    y='Unemployment Rate',
    color='Period',  # Use color to differentiate periods
    animation_frame='Period',  # Use 'Period' column for animation
    title='Estimated Unemployment Rate Before, During, and After Lockdown by State',

)
# Move the animation slider to the bottom
fig.update_layout(updatemenus=[dict(type='buttons', showactive=False, x=0.1, y=0.01)])

# Hide the legend
fig.update_layout(legend=dict(orientation="h", yanchor="bottom", y=1.02, xanchor="right", x=1))

# Update the Y axis Title
fig.update_yaxes(title_text="Unemployment Rate (%)")
fig.show()
```
